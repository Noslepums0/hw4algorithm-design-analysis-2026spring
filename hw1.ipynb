## 问题 1

根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ $\text{ for } n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ $\text{ for } n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ $\text{ for } n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ $\text{ for } n>1$
5. $T(n)=1 \text{ for } n \le 2;\; T(n)=T(\sqrt{n})+1 \text{ for } n>2$

### 问题 1 解答

下面默认把常数和取整影响吸收到渐近符号中。

1. $T(n)=T(n/4)+1$

递归深度满足 $n/4^k=1$，即 $k=\log_4 n$。每层代价为 $1$，所以渐进上界为$\log n$。

2. $T(n)=3T(n/3)+n^2$

由 Master 定理，$a=3,b=3$，$n^{\log_b a}=n$，而 $f(n)=n^2=\Omega(n^{1+\epsilon})$，且满足正则条件，所以渐进上界为$n^2$。

3. $T(n)=T(2n/3)+1$

递归深度满足 $n(2/3)^k=1$，即 $k=\log_{3/2} n$。每层代价为 $1$，所以渐进上界为 $\log n$。

4. $T(n)=5T(n/4)+n$

由 Master 定理，$a=5,b=4$，$n^{\log_4 5}\approx n^{1.161}$，而 $f(n)=n=O(n^{\log_4 5-\epsilon})$，所以渐进上界为$n^{\log_4 5}$。

5. $T(n)=T(\sqrt n)+1$

令 $n=2^m$，设 $S(m)=T(2^m)$，则  
$S(m)=S(m/2)+1$。  
递归深度为 $\log m=\log(\log n)$，所以渐进上界为$\log\log n$。

观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

```latex
DoSomething(A,p,r)

first call: DoSomething(A,1,n)
```

note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. \* 使用python编程实现上述算法或其对应的更高效版本的算法

### 问题 2 解答
1. 该算法每次递归处理长度约为 $\lceil 2n/3\rceil$ 的子数组，并调用 3 次，因此递归式为：

$$
T(n)=3T(\lceil 2n/3\rceil)+O(1)
$$

忽略取整，根据递归树：

$$
T(n)=\Theta\left(n^{\log_{3/2}3}\right)\approx \Theta(n^{2.71})
$$

2. 它的功能是把数组按非降序排序。基本思想是：先排序前 $2/3$，再排序后 $2/3$，最后再排序前 $2/3$，利用重叠区间逐渐纠正逆序。

它不是最高效的排序算法。比较排序中常用的归并排序、堆排序平均/最坏时间复杂度为 $O(n\log n)$，快速排序平均也是 $O(n\log n)$，都明显优于该算法的 $\Theta(n^{2.71})$。


````python
def do_something(A, p=0, r=None):
    
    if r is None:
        r = len(A) - 1

    n = r - p + 1
    if n <= 1:
        return A

    if n == 2 and A[p] > A[r]:
        A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = (2 * n + 2) // 3   # ceil(2n/3)
        do_something(A, p, p + m - 1)
        do_something(A, r - m + 1, r)
        do_something(A, p, p + m - 1)
    return A


# 测试
arr = [5, 1, 4, 2, 8, 3]
print(do_something(arr.copy()))

# 时间复杂度：T(n)=3T(ceil(2n/3))+O(1)=Theta(n^(log_{3/2}3))，约为 Theta(n^2.71)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

### 问题 3 解答

题目给出的profits[i]表示第 $i$ 天的收益变化。若从第 $l$ 天买入、第 $r$ 天卖出，则总收益等于区间profits[l..r]的和。因此问题等价于寻找最大连续子数组和，并记录区间左右端点。

可以用 Kadane 算法在线性时间内完成：

- 用current_sum表示“必须以当前天结尾”的最大收益；
- 如果之前的累计收益为负，则继续累加只会拖累当前结果，应从当前天重新开始；
- 同时维护全局最大收益best_sum及其对应起止日期。

时间复杂度：$O(n)$。  
空间复杂度：$O(1)$。


````python
def max_stock_profit_days(profits):
    if not profits:
        raise ValueError("profits 不能为空")

    best_sum = current_sum = profits[0]
    best_start = best_end = current_start = 0

    for i in range(1, len(profits)):
        if current_sum < 0:
            current_sum = profits[i]
            current_start = i
        else:
            current_sum += profits[i]

        if current_sum > best_sum:
            best_sum = current_sum
            best_start = current_start
            best_end = i

    return best_sum, best_start + 1, best_end + 1


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy_day, sell_day = max_stock_profit_days(profits)
print(f"第{buy_day}天买入，第{sell_day}天卖出，最大收益：{profit}")

# 时间复杂度：O(n)
# 空间复杂度：O(1)
````

## 问题 4

给定一个包含不同整数元素的数组 $A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素 $A[x]$ 为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在 $O(\log n)$ 的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

### 问题 4 解答

1. 因为数组元素互不相同，并且满足 $A[1]>A[2]$、$A[n-1]<A[n]$，所以数组内部一定存在局部最小值。

可以使用二分思想：

- 检查中点mid；
- 若 A[mid] 小于左右邻居，则它就是局部最小；
- 若 A[mid] > A[mid-1]，说明往左是下降方向，左边一定存在局部最小；
- 否则若 A[mid] > A[mid+1]，说明往右是下降方向，右边一定存在局部最小。

每次排除一半区间，因此时间复杂度为 $O(\log n)$，空间复杂度为 $O(1)$。



````python
def find_one_local_min(A):
    n = len(A)
    if n < 3:
        raise ValueError("数组长度至少为 3")

    left, right = 1, n - 2
    while left <= right:
        mid = (left + right) // 2

        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]

        if A[mid] > A[mid - 1]:
            right = mid - 1
        else:
            left = mid + 1

    return None

A = [9, 3, 7, 2, 1, 4, 5]
print("一个局部最小值：", find_one_local_min(A))


# find_one_local_min 时间复杂度：O(log n)，空间复杂度：O(1)

````

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

### 问题 5 解答

1. **先排序再输出前 k 个**

可以使用归并排序。排序总时间为 $O(n\log n)$，输出前 $k$ 个数为 $O(k)$，总时间复杂度为：

$$
O(n\log n+k)=O(n\log n)
$$

2. **不断插入建小根堆，再 Extract-min k 次**

若从空堆开始逐个插入 $n$ 个元素，每次插入代价为 $O(\log n)$，因此建堆时间复杂度为 $O(n\log n)$。  
之后执行 $k$ 次 Extract-min，每次 $O(\log n)$，总时间复杂度为：

$$
O(n\log n+k\log n)
$$


````python
import heapq

def merge_sort(arr):
    if len(arr) <= 1:
        return arr[:]

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    i = j = 0
    result = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def k_smallest_by_sort(arr, k):
    return merge_sort(arr)[:k]


def k_smallest_by_min_heap(arr, k):
    heap = []
    for x in arr:
        heapq.heappush(heap, x)  # 从空堆不断插入

    result = []
    for _ in range(min(k, len(heap))):
        result.append(heapq.heappop(heap))
    return result





arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(k_smallest_by_sort(arr, k))
print(k_smallest_by_min_heap(arr, k))


# 方法1时间复杂度：O(n log n)
# 方法2时间复杂度：O(n log n + k log n)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

### 问题 6 解答

只有当左边星球向右运动（正数），右边星球向左运动（负数）时，二者才会相撞。可以用栈模拟已经稳定保留下来的星球：

- 当前星球为正数，直接入栈；
- 当前星球为负数时，只可能与栈顶的正数相撞；
- 若栈顶正数质量更小，则栈顶消失，继续比较；
- 若栈顶正数质量更大，则当前负数消失；
- 若质量相同，根据题意：自右向左运动的负星球消失，自左向右运动的正星球保留，所以当前负数消失，栈顶保留。

每个星球最多入栈、出栈一次，因此时间复杂度为 $O(n)$，空间复杂度为 $O(n)$。


````python
def planet_collision(A):
    stack = []

    for planet in A:
        alive = True

        while alive and planet < 0 and stack and stack[-1] > 0:
            if stack[-1] < -planet:
                stack.pop()
                continue
            else:
                alive = False

        if alive:
            stack.append(planet)

    return stack


print(planet_collision([-3, -6, 2, 8, 5, -8, 9, -2, 1]))
print(planet_collision([23, -8, 9, -3, -7, 9, -23, 22]))

# 时间复杂度：O(n)
# 空间复杂度：O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

### 问题 7 解答

这是经典 0-1 背包问题。每块大理石只能选或不选，不能拆分。

设dp[i][w]表示只考虑前 i 件物品、容量为 w 时能获得的最大价值。对于第i 件物品，设重量为 weight，价值为 value：

- 不选它：价值为 dp[i-1][w]
- 选它：要求 w >= weight，价值为 dp[i-1][w-weight] + value

状态转移为：

$$
dp[i][w]=\max(dp[i-1][w], dp[i-1][w-weight]+value)
$$

时间复杂度：$O(nW)$。  
空间复杂度：$O(nW)$；若只求最大值，也可以优化为 $O(W)$。

本题中最大承重 $W=9$，最优选择为重量 4、价值 40 的大理石和重量 3、价值 50 的大理石，总重量 7，总价值 90。


````python
def knapsack_01(items, capacity):
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]
        for w in range(capacity + 1):
            dp[i][w] = dp[i - 1][w]
            if w >= weight:
                dp[i][w] = max(dp[i][w], dp[i - 1][w - weight] + value)

    chosen = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:
            chosen.append(i - 1)
            w -= items[i - 1][0]
    chosen.reverse()

    return dp[n][capacity], chosen


items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9
max_value, chosen = knapsack_01(items, capacity)
print("最大价值：", max_value)
print("选择的大理石：", [items[i] for i in chosen])

# 时间复杂度：O(nW)
# 空间复杂度：O(nW)
````

输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

### 问题 8 解答

二叉搜索树的中序遍历结果应当是严格递增序列。若恰好两个节点被交换，则中序序列中会出现 1 次或 2 次逆序。

算法步骤：

1. 对树做中序遍历；
2. 用 prev记录上一个访问的节点；
3. 若发现 prev.val > current.val：
   - 第一次发现时，记录 first = prev`，`second = current；
   - 若第二次发现，更新 second = current；
4. 遍历结束后交换 first.val 和 second.val。

时间复杂度：$O(n)$。  
空间复杂度：递归栈为 $O(h)$，其中 $h$ 是树高；最坏为 $O(n)$，平衡树为 $O(\log n)$。


````python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(level_list):
    if not level_list:
        return None

    root = TreeNode(level_list[0])
    queue = deque([root])
    i = 1

    while queue and i < len(level_list):
        node = queue.popleft()

        if i < len(level_list) and level_list[i] is not None:
            node.left = TreeNode(level_list[i])
            queue.append(node.left)
        i += 1

        if i < len(level_list) and level_list[i] is not None:
            node.right = TreeNode(level_list[i])
            queue.append(node.right)
        i += 1

    return root


def tree_to_level_list(root):
    if not root:
        return []

    result = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)

    while result and result[-1] is None:
        result.pop()
    return result


def recover_tree(root):
    first = second = prev = None

    def inorder(node):
        nonlocal first, second, prev
        if not node:
            return

        inorder(node.left)

        if prev and prev.val > node.val:
            if first is None:
                first = prev
            second = node
        prev = node

        inorder(node.right)

    inorder(root)

    if first and second:
        first.val, second.val = second.val, first.val

    return root


root1 = build_tree([1, 3, None, None, 2])
recover_tree(root1)
print(tree_to_level_list(root1))

root2 = build_tree([3, 1, 4, None, None, 2])
recover_tree(root2)
print(tree_to_level_list(root2))

# 时间复杂度：O(n)
# 空间复杂度：O(h)
````

### 问题 10 解答

从根节点到叶节点的路径可以看作逐位构造数字。若当前路径已经形成数字current，访问值为 x 的新节点后，新数字变为：

$$
current \times 10 + x
$$

因此可以用深度优先搜索：

- 从根节点开始，维护当前路径形成的数字；
- 到达叶节点时，把该数字加入答案；
- 对左右子树递归计算。

时间复杂度：$O(n)$，每个节点访问一次。  
空间复杂度：$O(h)$，其中 $h$ 是树高。


````python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(level_list):
    if not level_list:
        return None

    root = TreeNode(level_list[0])
    queue = deque([root])
    i = 1

    while queue and i < len(level_list):
        node = queue.popleft()

        if i < len(level_list) and level_list[i] is not None:
            node.left = TreeNode(level_list[i])
            queue.append(node.left)
        i += 1

        if i < len(level_list) and level_list[i] is not None:
            node.right = TreeNode(level_list[i])
            queue.append(node.right)
        i += 1

    return root


def sum_numbers(root):
    def dfs(node, current):
        if not node:
            return 0

        current = current * 10 + node.val

        if not node.left and not node.right:
            return current

        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


root1 = build_tree([1, 2, 3])
print(sum_numbers(root1))

root2 = build_tree([4, 9, 0, 5, 1])
print(sum_numbers(root2))

# 时间复杂度：O(n)
# 空间复杂度：O(h)
````
